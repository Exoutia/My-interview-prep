# Treats Distribution
Say we are given an integer array of an even length, where different numbers in the array
represent certain kinds of snacks or treats. Each number maps to, or represents, one kind of
snack. So the following array would have two kinds: snack type 3 and type 2:<br/>
```javascript 
const snacks = [3, 3, 2, 2]
```
![](../images/treats_distribution.png)
You need to distribute these snacks equally in number to a brother and sister duo. Write an
algorithm treatsDistribution(snacks: array) to return the maximum number of unique
kinds of snacks the sister could gain.
```javascript
const snacks = [2, 2, 3, 3, 4, 4];
ans: 3
```

## Algorithms
As we can see sister wants to find out the max number of unique treats she can get so we can just make a set of candies and return the result as len of set but theres a small brother also and she also has to share half of candies so we will only return half of array if unique elements is greater number cause she is good one chan.
<br/>
#### Using the sort
1. sort the number in array.
2. increase the count if we find a unique number which is not the same as the previous number.
3. if the count is equal to the half of len of array return the count.
<br/>
#### Using a set 
- Traverse over all the elements of the array
- Put the elements from the array into a Set
- As a set can only contain unique elements, the length of the set (uniqueCount) will be the number of unique elements
- Return the result as min(n/2, uniqueCount)
- The time complexity of using a Set is O(n).

````python
def count_unique_treats(arr):
    return min(len(arr)//2, len(set(arr)))

print(count_unique_treats([1,2,1,2,3,3]))
````
